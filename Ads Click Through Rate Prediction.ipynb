Chi Square Test

```python
import scipy.stats as stats
df=pd.crosstab(df['Gender'],df['Clicked on Ad'])
print(df)
```

```python
Observed_values = df.values
Observed_values
```

```python
val=stats.chi2_contingency(df)
val
```

```python
Expected_values = val[3]
Expected_values
```

```python
no_of_rows=len(df.iloc[0:,0])
no_of_columns=len(df.iloc[0,0:])
ddof=(no_of_rows-1)*(no_of_columns-1)
print("Degree of Freedom:-",ddof)
alpha = 0.05
```

```python
from scipy.stats import chi2
chi_square=sum([(o-e)**2./e for o,e in zip(Observed_values,Expected_values)])
chi_square_statistic=chi_square[0]+chi_square[1]
```

```python
print("chi-square statistic:-",chi_square_statistic)

critical_value=chi2.ppf(q=1-alpha,df=ddof)
print('critical_value:',critical_value)
```

```python
if chi_square_statistic>=critical_value:
    print("Reject H0,There is a relationship between 2 categorical variables")
else:
    print("Accept H0,There is no relationship between 2 categorical variables")
```

Overall, gender variations in potential ad clicks are minor.

# Model Selection

```python
dfm = df1.copy()
```

```python
dfm["Gender"] = dfm["Gender"].map({"Male": 0, "Female":1})
```

```python
X = dfm.drop(columns='Clicked on Ad').copy()  
Y = dfm['Clicked on Ad'].copy()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 4)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

Logistic Regression

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.linear_model import LogisticRegression
```

```python
model_lr = LogisticRegression()
model_lr.fit(X_train_scaled, Y_train)

# Make predictions on the test set
Y_pred_lr= model_lr.predict(X_test_scaled)

# Probability for Roc-Auc Curve
Y_pred_lr_proba = model_lr.predict_proba(X_test)

# Reshape
Y_test = Y_test.ravel()
Y_pred_lr_proba = Y_pred_lr_proba[:, 1]
```

```python
# Calculate evaluation metrics
accuracy = accuracy_score(Y_test, Y_pred_lr)
precision = precision_score(Y_test, Y_pred_lr)
recall = recall_score(Y_test, Y_pred_lr)
f1 = f1_score(Y_test, Y_pred_lr)

print("Accuracy:", round(accuracy*100,3))
print("Precision:", round(precision*100,3))
print("Recall:", round(recall*100,3))
print("F1 Score:", round(f1*100,3))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(Y_test, Y_pred_lr)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=model_lr.classes_)
disp.plot()
```
